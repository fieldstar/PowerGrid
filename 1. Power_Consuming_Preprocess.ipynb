# 数据预处理
### 对用户数据的奇异值进行删除
### 补充不连续的日期
### 对空数据进行填充

```python
tmp_df=pd.DataFrame() #进行用户数据的拼接
for i in user_num:
    cur_user_data=data[data['编号']==i]
    
    explorer=cur_user_data['日用电量'].describe()

    IRQ=explorer['75%']-explorer['25%']
    #过滤掉奇异值
    cur_user_data=cur_user_data[cur_user_data['日用电量']<=(explorer['75%']+IRQ*3)]
    
    tmp_date= cur_user_data['DATA_DATE']   #获取当前用户数据中的最早日期和最晚日期，以便于补充这两个日期之前的空数据
    
    first=min(tmp_date)
    last=max(tmp_date)
    
    cur_user_data=cur_user_data.set_index("DATA_DATE")  #当前用户设置DATA_DATE为数据框的索引
    #补足所缺的日期
    dt = pd.date_range(first, last) #生成一个日期序列
    #重新定义索引
    idx = pd.DatetimeIndex(dt) 
    cur_user_data =  cur_user_data.reindex(idx)  #对当前用户按照前面的日期序列重新索引后，原数据集不存在的日期将增加一个新的空行。
    
    #对空值按最近日期进行插值处理
    cur_user_data = cur_user_data.interpolate(method='time', axis=0).ffill().bfill() 
    
    cur_user_data['日期'] = cur_user_data.index #将索引列改为普通列
    cur_user_data.reset_index(inplace=True,drop=True) #重置索引，以避免下个用户的索引重复
    
    tmp_df=pd.concat([tmp_df,cur_user_data],ignore_index=True)
    
print(len(tmp_df))
```

# 按月度进行统计（减少粒度到月份）

```python
tmp_user_data_monthly=pd.DataFrame()
#按月度进行统计（减少粒度到月份）
for i in user_num:
    cur_user_data=cleaned_user_data[cleaned_user_data['编号']==i]
    monthly_groups = cur_user_data.resample('M')
    monthly_data = monthly_groups.sum()
    #占比不能直接相加，故重新计算
    days=monthly_data['编号']/i
    monthly_data['编号']=i
    monthly_data['尖峰用电量占比']=monthly_data['尖峰用电量']/monthly_data['日用电量']
    monthly_data['高峰用电量占比']=monthly_data['高峰用电量']/monthly_data['日用电量']
    monthly_data['平段用电量占比']=monthly_data['平段用电量']/monthly_data['日用电量']
    monthly_data['谷段用电量占比']=monthly_data['谷段用电量']/monthly_data['日用电量']
    
    monthly_data['尖峰日均用电量']=monthly_data['尖峰用电量']/days
    monthly_data['高峰日均用电量']=monthly_data['高峰用电量']/days
    monthly_data['平段日均用电量']=monthly_data['平段用电量']/days
    monthly_data['谷段日均用电量']=monthly_data['谷段用电量']/days

    
    monthly_data['尖峰平均负荷']=monthly_data['尖峰平均负荷']/days
    monthly_data['高峰平均负荷']=monthly_data['高峰平均负荷']/days
    monthly_data['平段平均负荷']=monthly_data['平段平均负荷']/days
    monthly_data['谷段平均负荷']=monthly_data['谷段平均负荷']/days

    tmp_user_data_monthly=pd.concat([tmp_user_data_monthly,monthly_data],ignore_index=False)
```

# 将用户月度数据输出到一个文件中保存

```python
#tmp_user_data_monthly.to_excel("CleanedPowerConsumingData_Monthly.xlsx",header=True)
tmp_user_data_monthly.to_csv("CleanedPowerConsumingData_Monthly.csv",header=True)
```
